```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
color_dict = {
    'Russian_Independent_Media':'#a6cee3',
    'Russian_Propaganda_Media':'#1f78b4',
    'Russian_Social_Media':'#b2df8a',
    'US_Social_Media':'#33a02c',
    'US_Big_Media':'#e31a1c',
    'US_Local_Media':'#fb9a99',
    'UK_Media':'#cab2d6',
    'Video_Hosting_Website':'#6a3d9a',
    'Short_URL':'#ff7f00',
    'unidentified':'#fdbf6f'
}
```

```python
df = pd.read_csv('agg_new_data.csv', parse_dates={'date':['tweet_time','tweet_time.1','tweet_time.2']})

labels = df.columns[1:-1]

values_list = []

for i in df.columns[1:-1]:
    x = df[i]
    values_list.append(x)

x = df.date
y1 = values_list[0]
y2 = values_list[1]
y3 = values_list[2]
y4 = values_list[3]
y5 = values_list[4]
y6 = values_list[5]
y7 = values_list[6]
y8 = values_list[7]
y9 = values_list[8]
y10 = values_list[9]
y11 = values_list[10]

plt.figure(figsize=(20,10))
plt.stackplot(x, y1, y2, y3, y4, y5, y6, y7, y8, y9, y10, y11, labels = labels, baseline='sym', colors=[color_dict.get(x) for x in df.columns[1:-1]])
plt.legend(loc='upper left')
plt.xlabel('Date')
plt.ylabel('Number of tweets')
plt.title('Themeriver')
plt.ylim(-5000,5000)
plt.show()
```
