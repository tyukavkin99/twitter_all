```python
new_df = pd.read_csv('category_data.csv', parse_dates=['tweet_time'])
```

```python
new_df
```

```python
dates_data_month = new_df.groupby([new_df.tweet_time.dt.year, new_df.tweet_time.dt.month, 'category'])['new_urls'].count()
```

```python
dates_df_month = dates_data_month.to_frame()
dates_df_month.reset_index(level='category', inplace=True)
dates_df_month.columns = ['category', 'count']
pivot_df_month = dates_df_month.pivot(columns = 'category', values = 'count')
pivot_df_month = pivot_df_month.fillna(0)
pivot_df_month
```

```python
pivot_df_month.plot(figsize=(20,10))
```

```python
pivot_df_month.to_csv('all_data_month.csv')
```
